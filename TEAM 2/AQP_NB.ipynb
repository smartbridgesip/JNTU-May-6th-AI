```python
from keras import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

## INPUT LAYER

```python
model.add(Dense(20,input_dim=4))
```

## HIDDEN LAYER

```python
model.add(Dense(20))
```

## OUTPUT LAYER

```python
model.add(Dense(1))
```

## COMPILE

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
```

```python
def model_reg():
    model=Sequential()
    model.add(Dense(20,input_dim=4))
    model.add(Dense(20))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
reg = KerasRegressor(build_fn=model_reg,batch_size=3,epochs=100)
```

```python
result = reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
y_pred
```

```python
mn.inverse_transform(y_pred.reshape(1,-1))
```
