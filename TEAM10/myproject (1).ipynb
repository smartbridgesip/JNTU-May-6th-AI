```python
x_train,x_test,y_train,y_test=train_test_split(c,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train) #finding out the best value and transforming or applying train set
x_test=sc.transform(x_test) 
y_train=sc.fit_transform(y_train) #finding out the best value and transforming or applying train set
y_test=sc.transform(y_test) #here we find the best value which we found out in x_train which is sc  will be modified which we don't want
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
c.shape
```

```python
model.add(Dense(input_dim=5,init="random_uniform",activation='relu',output_dim=15))
```

```python
#hidden layer
model.add(Dense(32))
```

```python
#output 
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
```

```python
def model_reg():
    model=Sequential()
    model.add(Dense(10,input_dim=5))
    model.add(Dense(20))
    model.add(Dense(1))
    model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
reg=KerasRegressor(build_fn=model_reg,batch_size=12,epochs=100)
```

```python
result=reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```
